```python
def preprocessing(txt):
  txt=txt.lower()
  txt=re.sub('[^a-z\s]',' ',txt)
  txt=re.sub(r'\s+', ' ', txt)
  return txt.strip()
```

```python
tweets['text']=tweets['text'].apply(preprocessing)
```
