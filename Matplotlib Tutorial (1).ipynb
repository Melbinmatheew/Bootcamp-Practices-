# Matplotlib

Matplotlib is a low level graph plotting library in python that serves as a visualization utility.

Matplotlib is open source and we can use it freely.

Most of the Matplotlib utilities lies under the pyplot submodule, and are usually imported under the plt alias:
    matplotlib.pyplot as plt
    
Plotting x and y points
The plot() function is used to draw points (markers) in a diagram.

By default, the plot() function draws a line from point to point.

The function takes parameters for specifying points in the diagram.

Parameter 1 is an array containing the points on the x-axis.

Parameter 2 is an array containing the points on the y-axis.

The x-axis is the horizontal axis.

The y-axis is the vertical axis.

#Markers

You can use the keyword argument marker to emphasize each point with a specified marker:
    
        ...
        
        plt.plot(ypoints, marker = '*')
        
                                    ...
                                    
#Format Strings fmt

You can also use the shortcut string notation parameter to specify the marker.

This parameter is also called fmt, and is written with this syntax:

marker|line|color

        ...
        
        ypoints = np.array([3, 8, 1, 10])

        plt.plot(ypoints, 'o:r')
        
        plt.show()
        
                                ...
                                
#Marker Size

You can use the keyword argument markersize or the shorter version, ms to set the size of the markers:

        ...
        
        ypoints = np.array([3, 8, 1, 10])

        plt.plot(ypoints, marker = 'o', ms = 20)
        
        plt.show()
        
                        ...
                        
#Marker Color

You can use the keyword argument markeredgecolor or the shorter mec to set the color of the edge of the markers:

        ...
        
        ypoints = np.array([3, 8, 1, 10])

        plt.plot(ypoints, marker = 'o', ms = 20, mec = 'r')
        
        plt.show()
        
                        ...

# Matplotlib Line

#Linestyle

You can use the keyword argument linestyle, or shorter ls, to change the style of the plotted line:

#Shorter Syntax

The line style can be written in a shorter syntax:

linestyle can be written as ls.

dotted can be written as :.

dashed can be written as --.

        ...
        
        plt.plot(ypoints, ls = ':')
        
        ...
        
#Line Color

You can use the keyword argument color or the shorter c to set the color of the line:
    
            ...
        
            plt.plot(ypoints, c = '#4CAF50')
            
            ...
            
#Line Width

You can use the keyword argument linewidth or the shorter lw to change the width of the line.

The value is a floating number, in points:
    
            
            ...
            
            plt.plot(ypoints, linewidth = '20.5')
            
            ...
            
#Multiple Lines

You can plot as many lines as you like by simply adding more plt.plot() functions:
    
            ...
        
            import matplotlib.pyplot as plt
            
            import numpy as np

            y1 = np.array([3, 8, 1, 10])
            
            y2 = np.array([6, 2, 7, 11])

            plt.plot(y1)
            
            plt.plot(y2)

            plt.show()
            
            ...
            

You can also plot many lines by adding the points for the x- and y-axis for each line in the same plt.plot() function.

(In the examples above we only specified the points on the y-axis, meaning that the points on the x-axis got the the default values (0, 1, 2, 3).)

The x- and y- values come in pairs:
    
            ...
        
            x1 = np.array([0, 1, 2, 3])
            
            y1 = np.array([3, 8, 1, 10])
            
            x2 = np.array([0, 1, 2, 3])
            
            y2 = np.array([6, 2, 7, 11])

            plt.plot(x1, y1, x2, y2)
            
            plt.show()
            
            ...

Line plots are typically used more commonly with regression and time series data rather than with classification or clustering data. Here's how line plots are typically used in each context:

◈Regression:

In regression analysis, line plots are often used to visualize the relationship between the independent variable(s) and the dependent variable.
The line in a regression plot represents the best-fit line or curve that describes the relationship between the variables.
Line plots help in understanding how changes in the independent variable(s) affect the dependent variable.
While line plots are not the only visualization tool used in regression analysis, they are commonly employed to provide a visual representation of the regression model's predictions and how they align with the actual data points.

◈Time Series Data:

Line plots are extensively used in time series analysis, where data is collected at regular intervals over time.
Time series data often includes measurements such as stock prices, temperature readings, sales figures, etc.
Line plots are used to visualize trends, patterns, and seasonality in the time series data.
They help in identifying long-term trends, cyclical patterns, and any irregular fluctuations in the data over time.
Time series forecasting techniques often rely on line plots to visualize historical data and model predictions.

◈Classification and Clustering:

Line plots are less commonly used with classification and clustering data compared to regression and time series data.
In classification tasks, line plots may not be as useful because the focus is on predicting discrete classes rather than modeling continuous relationships.
In clustering, line plots may occasionally be used to visualize certain aspects of the data or results, but they are not a primary visualization tool for clustering algorithms.

While line plots can provide insights into the relationships and patterns present in the data, their usage tends to be more prevalent in regression analysis and time series analysis due to their suitability for visualizing continuous variables and trends over time.

# Matplotlib Subplot`

## Display Multiple Plots

### With the subplot() function you can draw multiple plots in one figure:

```python
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)

plt.show()
```

 The subplot() Function
    
The subplot() function takes three arguments that describes the layout of the figure.

The layout is organized in rows and columns, which are represented by the first and second argument.

The third argument represents the index of the current plot.

plt.subplot(1, 2, 1)    #the figure has 1 row, 2 columns, and this plot is the first plot.

plt.subplot(1, 2, 2)    #the figure has 1 row, 2 columns, and this plot is the second plot.

So, if we want a figure with 2 rows an 1 column (meaning that the two plots will be displayed on top of each other instead of side-by-side), we can write the syntax like this:

```python
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 1, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 1, 2)
plt.plot(x,y)

plt.show()
```

# Scatter plot

In a scatter plot, the data points are represented individually using dots or circles. This type of plot is generally used to represent the relationship between two variables.

Scatter plots are commonly used in both regression and clustering analysis, but they are not typically used in classification tasks. Here's how scatter plots are utilized in each context:

◈Regression:

In regression analysis, scatter plots are frequently used to visualize the relationship between two continuous variables.
Each point on the scatter plot represents a pair of values for the independent and dependent variables.
Scatter plots help to identify patterns, trends, and the strength of the relationship between variables.
They are often used in conjunction with regression lines or curves to visually assess the fit of the regression model to the data.
Outliers and influential data points can also be identified on scatter plots, which may influence the regression analysis.

◈Clustering:

Scatter plots can also be useful in clustering analysis, particularly for visualizing the results of clustering algorithms.
In clustering, each point in the scatter plot represents an individual data point, and its position is determined by its feature values.
Scatter plots can help in assessing the separation or overlap between clusters and identifying any inherent patterns or structures in the data.
While scatter plots are useful for visualizing clusters when the data has two or three dimensions, additional techniques such as dimensionality reduction or clustering-specific visualization methods are often used for higher-dimensional data.

◈Classification:

Scatter plots are not typically used in classification tasks because classification involves predicting categorical labels rather than analyzing continuous relationships between variables.
In classification tasks, visualizations such as confusion matrices, ROC curves, or precision-recall curves are more commonly used to assess the performance of classification models and understand the distribution of predicted classes.


Overall, scatter plots are versatile visualizations that are particularly useful in regression analysis for understanding relationships between continuous variables and in clustering analysis for visualizing clusters and patterns in the data.




Size

You can change the size of the dots with the s argument.

Just like colors, make sure the array for sizes has the same length as the arrays for the x- and y-axis:

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes)

plt.show()
```

Combine Color Size and Alpha

You can combine a colormap with different sizes of the dots. This is best visualized if the dots are transparent:

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')

plt.colorbar()

plt.show()
```

### Bar graph

##### A bar graph is used to compare categorical data, where the height of the bars represents its value.

```python
#Let’s compare the number of people who survived and those who did not:

survivor = data['survived'].value_counts()

plt.bar(x=[0,1], height=survivor.values, width=0.2, color='yellow', edgecolor='red',linewidth=2);
print('Thus the majority didn’t survive.')
```

A bar plot is primarily used for visualization purposes and is not inherently tied to any specific machine learning task such as classification, regression, or clustering. However, it can be used in various stages of these tasks for different purposes:

◈Classification:

Visualizing class distributions: Bar plots can be used to visualize the distribution of different classes in the dataset, which can provide insights into class imbalances.
Visualizing performance metrics: Bar plots can be used to display performance metrics such as accuracy, precision, recall, and F1-score for different classifiers.

◈Regression:

Visualizing feature importance: Bar plots can be used to show the importance of different features in regression models, such as coefficients in linear regression or feature importance scores in tree-based models.
Visualizing actual vs. predicted values: Bar plots can be used to compare actual and predicted values for different instances or groups, providing insights into model performance.

◈Clustering:

Visualizing cluster distributions: Bar plots can be used to visualize the distribution of data points across different clusters.
Visualizing cluster characteristics: Bar plots can be used to display the mean or median values of different features within each cluster, helping to interpret the characteristics of the clusters.

    In summary, while bar plots themselves are not directly related to classification, regression, or clustering tasks, they can be used as a tool for visualizing data and results at various stages of these tasks.

# Matplotlib Histograms

### Histogram

A histogram is a graph showing frequency distributions.

It is a graph showing the number of observations within each given interval.



A histogram is a bar graph-like plot used to represent the frequency distribution of a single variable. The whole range of frequencies is divided into equal units called ‘bins’.

```python
age=data['age']
plt.hist(age);
```

Histograms are primarily used for data visualization and exploration, rather than being directly tied to specific machine learning tasks like classification, regression, or clustering. However, they can be helpful in understanding the distribution of data, which can be useful in various stages of these tasks:

◈Classification:

Data exploration: Histograms can help visualize the distribution of features within different classes, aiding in feature selection and understanding class separability.
Class imbalance detection: Histograms can highlight any class imbalances within the dataset, which is important to address during classification model training.

◈Regression:

Feature exploration: Histograms can assist in understanding the distribution of independent variables (features) and their relationship with the target variable in regression tasks.
Residual analysis: In regression analysis, histograms of residuals (the differences between actual and predicted values) can help diagnose model adequacy and identify potential issues like heteroscedasticity.

◈Clustering:

Data distribution exploration: Histograms can provide insights into the distribution of features within the dataset, which can be helpful in understanding the structure of the data before performing clustering.
Cluster assessment: Histograms can be used to examine the distribution of data points within identified clusters, aiding in cluster interpretation and validation.

        In summary, while histograms themselves are not directly associated with any specific machine learning task, they serve as valuable tools for data exploration and understanding, which are essential in the preprocessing, analysis, and interpretation stages of classification, regression, and clustering tasks.

# Matplotlib Pie Charts


### Pie chart
A pie chart is a circle divided into slices, where each slice represents the different categories under that variable. The size of the slices depends on the relative percentage or count of the categories.

Function used - plt.pie()

```python
plt.pie(data['sex'].value_counts(),labels=['Male','Female']);
```

```python
plt.pie(data['survived'].value_counts(),labels =['Not Survived','Survived'],startangle=90,
       autopct='%.1f%%');
```


Pie plots, like histograms, are primarily used for data visualization and exploration rather than being directly tied to specific machine learning tasks such as classification, regression, or clustering. However, they can still have applications in these tasks:

◈Classification:

Class distribution visualization: Pie plots can be used to visualize the distribution of classes in a dataset, helping to identify class imbalances or understand the relative proportions of different classes.

◈Regression:

Categorical feature distribution: If there are categorical variables in the dataset, pie plots can be used to visualize their distribution, which can be useful for understanding relationships between categorical variables and the target variable in regression tasks.

◈Clustering:

Cluster distribution visualization: Pie plots can be used to visualize the distribution of data points across different clusters, providing insights into the sizes or proportions of clusters identified by clustering algorithms.

        While pie plots themselves are not directly associated with these machine learning tasks, they can be used as supplementary tools for data exploration and visualization, aiding in the understanding of data characteristics and patterns, which is crucial in the context of classification, regression, and clustering tasks.
