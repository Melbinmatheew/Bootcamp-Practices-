```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv(r"D:\Data Set\medical_insurance.csv")
data.head()
```

```python
n=['sex','smoker','region']
for i in n:
    print(i,data[i].unique())
    print()
```

```python
from sklearn import preprocessing
label=preprocessing.LabelEncoder()
v=['sex','smoker','region']
for i in v:
    data[i]=label.fit_transform(data[i])
```

```python
data.head()
```

```python
data.info()
```

```python
X=data.drop(columns='charges', axis=1)
y=data['charges']
```

```python
from sklearn.model_selection import train_test_split
Xtrain,Xtest,ytrain,ytest=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression,Ridge
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
```

```python
models={
    "Decision Tree Regression":DecisionTreeRegressor(),
    "Multiple Linear Regression":LinearRegression(),
    "Ridge regression":Ridge()
}
```

```python
print("Regression Model Evaluation")
print()
for name,model in models.items():
    model.fit(Xtrain,ytrain)
    predictions=model.predict(Xtest)
    r2=r2_score(ytest,predictions)
    mse=mean_squared_error(ytest,predictions)
    mae=mean_absolute_error(ytest,predictions)
    
    print("Model :",name)
    print("R2_score :",r2)
    print('Mean Squared Error :',mse)
    print('Mean Absolute Error :',mae)
    print()
```

```python
import numpy as np

def predict_insurance_charges():
    # Input from the user
    age = int(input("Enter age: "))
    sex = input("Enter sex (male/female): ").lower()
    bmi = float(input("Enter BMI: "))
    children = int(input("Enter number of children: "))
    smoker = input("Is the person a smoker (yes/no): ").lower()
    region = input("Enter region (southwest/southeast/northwest/northeast): ").lower()

    # Convert categorical inputs to numerical labels
    sex_label = 0 if sex == 'female' else 1
    smoker_label = 1 if smoker == 'yes' else 0
    region_label = {'southwest': 0, 'southeast': 1, 'northwest': 2, 'northeast': 3}[region]

    # Create input array
    input_data = np.array([[age, sex_label, bmi, children, smoker_label, region_label]])

    # Use Decision Tree Regression model for prediction
    selected_model = models["Decision Tree Regression"]

    # Predict charges using the selected model
    predicted_charges = selected_model.predict(input_data)

    print("Predicted insurance charges:", predicted_charges[0])

# Call the function
predict_insurance_charges()
```
